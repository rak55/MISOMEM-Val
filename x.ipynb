```python
def processCycles(resources, conversionRate):

    cyc = 0
    
    while True:
        cp = resources.count('P')
        ca = resources.count('A')
        
        if cp >= conversionRate:
            deleted = 0
            for i in range(len(resources) - 1, -1, -1):
                if resources[i] == 'P':
                    resources.pop(i)
                    deleted += 1
                    if deleted == conversionRate:
                        break
            resources.insert(0, 'A')
            cyc += 1
        
        elif ca > 0:
            for i in range(len(resources) - 1, -1, -1):
                if resources[i] == 'A':
                    resources[i] = 'P'
                    break
            cyc += 1
        
        else:
            break
    
    return cyc
```

```python
def formatNewspaper(paragraphs, width):
    result = []

    def add_line(line):
        total_len = len(line)
        extra = width - total_len
        if extra % 2 == 0:
            left = right = extra // 2
        else:
            left = extra // 2
            right = extra // 2 + 1
        centered = " " * left + line + " " * right
        result.append(f"*{centered}*")

    for para in paragraphs:
        line = ""
        for chunk in para:
            if not line:
                line = chunk
            elif len(line) + 1 + len(chunk) <= width:
                line += " " + chunk
            else:
                add_line(line)
                line = chunk
        if line:
            add_line(line)

    border = "*" * (width + 2)
    return [border] + result + [border]
```
